```python
##importing data sets from a flat file to pandas DataFrame

#importing pandas as pd
import pandas as pd

#read CSV file "world_population.csv" and store is as a DataFrame called df
df = pd.read_csv ('world_population.csv', index_col=0)

#print df
print(df)
```

```python
##INDEXING AND SUBSETTING

#Implement multi_level indexes on column Continent, Country/Territory, and Capital from df as the indexes using .set_index()
df_ind = df.set_index(["Continent","Country/Territory", "Capital"])
print(df_ind)
```

```python
##Indexes make subsetting easier

#subset Asia continent as most largest World Population Percentage using .loc[]                                                                     
df_ind.loc[["Asia"]]
```

```python
##GROUPING AND SUMMARIZING DATA

#Use Pandas .groupby() to Group the continent then sum 2022 Population, Area (km²), Growth Rate and  World Population Percentage using .sum(), store in continent_totals

continent_totals_pop = df.groupby("Continent")[["2022 Population", "Area (km²)", "Growth Rate","World Population Percentage" ]].sum()
print(continent_totals_pop)
```

```python
##SORTING DATA

#Sort continent_totals_pop by the values of 2022 Population from highest to lowest using .sort_values()
#Create continent_totals_pop, storing as continent_totals_pop_sorted
continent_totals_pop_sorted = continent_totals_pop.sort_values(["2022 Population"], ascending = False)

#print continent_totals_pop_sorted
print(continent_totals_pop_sorted)
```

```python
#To know the population density of each continent, I divide the number of 2022 population 
#by the measurement of the area
pop_density_2022 = continent_totals_pop["2022 Population"]/continent_totals_pop["Area (km²)"]  
print(pop_density_2022)
```

```python
#JOINING DATAFRAMES

##Joining dataframes with merging methode. Merging type is left join using .merge()
#importing data sets from a flat file to pandas DataFrame


#read CSV file "world-population-by-country-2020.csv" and store is as a DataFrame called df_2
df_2 = pd.read_csv ('world-population-by-country-2020.csv', index_col=0)

#print df_2
print(df_2)
```

```python
df_3 = df.merge(df_2, left_on ='Country/Territory',right_on= 'Country (or dependency)', how='left')
print(df_3.head())
```

```python
#to get to know the countries with the most highest density, I use .sort_value() with descending order.
df_3.sort_values(["Density  (P/Km²)"], ascending = False)
```

```python
#it returns that Malaysia is the highest density, following by Dominica, Syria, Cambodia, Kenya.
```

# Data Visualisation

```python
#Data Visualisation with Matplotlib and Seaborn
#Import the matplotlib.pyplot submodule and name it plt and seaborn as sns 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
##Data visualisation using Matplotlib 
# Create a Figure and an Axes with plt.subplots
fig, ax = plt.subplots()

fig, ax = plt.subplots()
ax.plot(pop_density_2022.index,pop_density_2022)

# Set the x-axis label
ax.set_xlabel('Continent')

# Set the y-axis label
ax.set_ylabel('Population Density 2022 (*1000000)')

# Show the figure
plt.show()
```

```python
#Data visualisation using Seaborn
sns.relplot(x=pop_density_2022, y="Continent", data=pop_density_2022, kind="scatter",size="Continent", hue="Continent")
```

```python
#Web scraping using BeautifulSoup
from bs4 import BeautifulSoup
import requests

url = "https://www.donedeal.ie/cars?make=Fiat;model:500,500C,500L,500X,500e&area=Dublin&fuelType=Petrol&transmission=Manual&year_from=2009&price_to=7000"
page = requests.get(url)

html_doc = page.text

soup = BeautifulSoup(html_doc)
pretty_soup = soup.prettify()

print(pretty_soup)
```

```python
#to get the title of the web using soup.title
car_title = soup.title

print(car_title)
```

```python
a_tags = soup.find_all()
for link in soup.find_all('a'):
    print(link.get('href'))
```
